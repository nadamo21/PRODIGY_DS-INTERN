```python
# Select the country data
country = df[df['Country Name'] == 'Egypt, Arab Rep.']

# Extract years and population data, ignoring non-numeric columns
years = country.columns[1:]  # Skip the 'Country Name' column
years = [year for year in years if year.isdigit()]  # Keep only numeric years
population = country[years].values.flatten()  # Get population data

# Convert years to integers
years = list(map(int, years))

# Plotting
plt.figure(figsize=(13, 10))
sns.barplot(x=years, y=population)
plt.xlabel('Years')
plt.ylabel('Population')
plt.title('Population of Egypt per year')
plt.xticks(rotation=90)
plt.show()
```

```python
years = df.columns[1:]  # Assuming the first column is not a year
years = [year for year in years if year.isdigit()]  

# Calculate the total population for each year based on filtered years
total_population = df[years].sum()

# Convert years to integers (if they are not already)
years = list(map(int, years))

# Verify that the lengths match
print(f"Length of years: {len(years)}, Length of total_population: {len(total_population)}")

# Plotting
plt.figure(figsize=(20, 10))  
sns.barplot(x=years, y=total_population)
plt.title('Total Population of the World per Year', fontsize=13)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Population', fontsize=12)
plt.xticks(rotation=90)
plt.show()
```

# Select top 10 countries with highest female population


```python
female_population_data = df[df['Indicator Code']=='SP.POP.TOTL.FE.IN']
female_population_sorted =female_population_data.sort_values(by="2023",ascending=False)
female_top_ten_countries = female_population_sorted .head(10)
print("Top ten countries of female population")
print(female_top_ten_countries[['Country Name']])
```

```python
male_top_ten_countries = pd.DataFrame({
    'Country Name': ['Low & middle income', 'IBRD only', 'Early-demographic dividend', 'Lower middle income', 'East Asia & Pacific', 'Egypt, Arab Rep.', 'South Asia', 'Country H', 'IDA total', 'OECD members'],
    '2023': [50, 45, 40, 35, 30, 25, 20, 15, 10, 5]
})

female_top_ten_countries = pd.DataFrame({
     'Country Name': ['Low & middle income', 'IBRD only', 'Early-demographic dividend', 'Lower middle income', 'East Asia & Pacific', 'Egypt, Arab Rep.', 'South Asia', 'Country H', 'IDA total', 'OECD members'],
    '2016': [45, 40, 35, 30, 25, 20, 15, 10, 5, 0]
})

top_ten_countries = pd.merge(
    male_top_ten_countries[['Country Name', '2023']],
    female_top_ten_countries[['Country Name', '2016']],
    on='Country Name',
    how='inner'
)

top_ten_countries = top_ten_countries.rename(columns={'2023': 'Male Population', '2016': 'Female Population'})
fig, ax = plt.subplots(figsize=(15, 8))
bar_width = 0.35
index = range(len(top_ten_countries))
ax.bar(index, top_ten_countries['Male Population'], bar_width, label='Male Population', color='blue')
ax.bar([i + bar_width for i in index], top_ten_countries['Female Population'], bar_width, label='Female Population', color='red')
ax.set_xlabel('Country')
ax.set_ylabel('Population')
ax.set_title('Top Ten Countries with Highest Male and Female Population (2023)')
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(top_ten_countries['Country Name'], rotation=90)
ax.legend()

plt.tight_layout()
plt.show()
```
